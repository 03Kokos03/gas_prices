```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
gas.corr()
```

```python
import seaborn as sns
```

```python
matrix = gas.corr().round(2)
plt.figure(figsize=(16,11))
plt.title('Gas correlation')

mask = np.triu(np.ones_like(matrix, dtype=bool))
sns.heatmap(matrix, annot=False , cmap="coolwarm", mask=mask)
plt.savefig('gas_corr.png')
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))
plt.title('Gas prices')
plt.plot(gas.Year , gas.USA , label='USA')
plt.plot(gas.Year , gas.Canada , label='Canada')
plt.plot(gas.Year , gas.UK , label='UK')
plt.xticks([1990,1994,1998,2002,2005,2008])

plt.legend()

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')


plt.title('Gas prices')
for country in gas:
  if country != 'Year':
    plt.plot(gas.Year , gas[country] , marker='.', label=country)
    
plt.xlabel('Year')
plt.ylabel('USD')    
plt.legend(loc='upper left',labelspacing=0.1)  

plt.savefig('gas_price.png')

plt.show()
```
